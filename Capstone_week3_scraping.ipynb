```python
from bs4 import BeautifulSoup # to parse the data returned from the website 
import requests as rs #to handle requests and query page
import pandas as pd #to create an dataframe
!conda install -c anaconda lxml --yes
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

#to make a query to link
source = rs.get(link).text
#print(source)
soup = BeautifulSoup(source,'lxml')
#print(soup)
table = soup.table
print(table)
```

```python
# Converting a HTML table to dataframe
Toronto_Neigh = pd.read_html(str(table), header=0)[0]
Toronto_Neigh.head(10)
#Toronto_Neigh.shape
#Toronto_Neigh.columns
```

```python
# Dropping rows that have Borough as Not assigned
for index,row in Toronto_Neigh.iterrows():
    if row['Borough']=="Not assigned":
        Toronto_Neigh = Toronto_Neigh.drop(index)
Toronto_Neigh.head()
#Toronto_Neigh.shape
```

```python
# Assigning the borough value to neighbouhood in case it is Not assigned while borough has a value
for index,row in Toronto_Neigh.iterrows():
    if row['Neighbourhood']=="Not assigned":
        row['Neighbourhood'] = row['Borough']
Toronto_Neigh.head()
#Toronto_Neigh.shape
```

```python
# Grouping the dataframe 
Toronto_Neigh = Toronto_Neigh.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join)
#    Toronto_Neigh['Neighbourhood'].apply(','.join)
type(Toronto_Neigh)
```

```python
# converting the Pandas series to dataframe and reseting the index to get the required result
Toronto_Neigh = pd.DataFrame(Toronto_Neigh).reset_index()
```
